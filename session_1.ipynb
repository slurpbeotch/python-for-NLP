```python
##### Web Scrapping
from urllib.request import urlopen
url = "https://simple.wikipedia.org/wiki/Venus"
page = urlopen(url)
print(page)
html_text = page.read()
print(html_text)
type(html_text)
```

```python
html_text = html_text.decode("utf-8")
start_index = html_text.find('<title>')
start_index
```
